```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
poke = pd.read_csv("C:/Users/marwi/OneDrive/Desktop/t_direct/pokemon.csv")
poke.head()
```

```python
plt.figure(figsize=(15, 5))

plt.scatter(x=poke['Type 1'], y=poke['Attack'], color='green')
plt.xlabel('Element Types')
plt.ylabel('Attack Power')
plt.title('Power Based on Element Type')

plt.show();
```

Here is a scatterplot of attack power based on the element type of each pokemon. We can see the range of attack strength and whether the average indexes of each type are known to be stronger or weaker. For example, the psychic type has some of the highest marks for attack power but the bar for the rest of the pokemon is in the lower range compared to other types.

```python
plt.figure(figsize=(15, 5))
plt.plot(poke.groupby('Type 1')['Attack'].mean());
```

This plot uses the groupby method for the element type and attack power to find the mean attack for each element. As you can see the psychic type has some of the strongest but because of the rest of the pokemon are usually weaker, it is average for a psychic type to be weak. Here we can see clearly on average the dragon type is on average the strongest type.

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(poke['Attack'], color='green',)
plt.plot(poke['Sp. Atk'], color='purple')
plt.xlabel('PokeDex')
plt.ylabel('Attack Power')

plt.subplot(1, 2, 2)
plt.plot(poke['Defense'], color='blue')
plt.plot(poke['Sp. Def'], color='orange')
plt.xlabel('PokeDex')
plt.ylabel('Defense Power')

plt.tight_layout()
plt.show()
```

Here are two subplots with two series in each. These are comparing the attack power with the special attack power for each pokemon as well the defense and special defense. This shows that just because a pokemon has high attack power, it does not necessarily correlate with their special attack power and similarly in comparison with defense and special defense. 

```python
plt.figure(figsize=(15, 5))

plt.plot(poke.groupby('Type 1')['Speed'].mean());
plt.xlabel('Element Types')
plt.ylabel('Speed')
plt.title('Which Element is the Fastest?')

plt.tight_layout()
plt.show();
```

This plot is used to see which elements are the fastest. If you are making a decision on your first pokemon based on attack and speed, then using this with the plot based on strength will help. This graph shows that the flying type is the fastest, but not too far behind we have the electric, dragon, dark, and fire which leaves plenty of options but of course this does not make catching your ideal pokemon any easier.
